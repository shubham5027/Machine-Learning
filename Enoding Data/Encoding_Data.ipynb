```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder #input non-order
from sklearn.preprocessing import LabelEncoder  #output label
from sklearn.preprocessing import OrdinalEncoder #input order
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('customer.csv')
```

```python
df
```

```python
X=df.drop(['purchased','age','gender'],axis=1)
Y=df['purchased']
X.sample(4),Y.sample(4)
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train
```

```python
le=LabelEncoder() #output_data
```

```python
y_train_le=le.fit_transform(y_train)
y_test_le=le.transform(y_test)
```
